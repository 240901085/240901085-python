```python
import pandas as pd

# 1. Define the employee data
data = {
    'Employee_ID': [101, 102, 103, 104],
    'Name': ['Alice Smith', 'Bob Jones', 'Charlie Brown', 'Daisy Miller'],
    'Department': ['HR', 'IT', 'Finance', 'Sales'],
    'Salary':[40000,67000,87900,45999]

}

# 2. Convert to DataFrame and Save
# We use index=False to keep the CSV clean
df_initial = pd.DataFrame(data)
file_name = "employees details - employee details.csv"
df_initial.to_csv(file_name, index=False)

# 3. Read the file back into variable 'd'
d = pd.read_csv(file_name)

# 4. Display the loaded data
print(d)
```
