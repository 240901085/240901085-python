```
import pandas as pd
import numpy as np
arr=np.array(['P','a','n','d','a','s'])
a=pd.Series(arr)
print("Series from array")
print(a)
```

```
arr={'x':0.,'y':1.,'z':2.,}
b=pd.Series(arr)
print("\nSeries from dictionary:\n")
print(b)
```

```
x=pd.Series(14,index=[0,1,2,3])
print("\nSeries using scalar\n")
print(x)
```

```
x=pd.Series([1,2,3],index=['a','b','c'])
print("Series through index:")
print(x)
```

```
a=pd.Series(data=[1,2,3,4])
print("\n a series:\n",a)
print("\nIndex:\n",a.index)
print("\nValues:\n",a.values)
print("\nShape:",a.shape)
print("\nDimension:",a.ndim)
print("\nSize:",a.size)
```
